```python
#import data set
import pandas as pd
columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'target']
iris = pd.read_csv('iris.csv', names=columns)
#for some reason the last row is NaN so we remove that
iris = iris[:len(iris)-1]
iris
```

```python
#clean up data and apply labels
features = ['sepal length', 'sepal width', 'petal length', 'petal width']
target = 'target'
X = iris[features].values
y = iris[target].values
#convert y to integers
y[y=='Iris-setosa']=0
y[y=='Iris-versicolor']=1
y[y=='Iris-virginica']=2
y = y.astype(int)
```

```python
#divide data into 5
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#train, test, and determine error 
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(3).fit(X_train,y_train)
knn
```

```python
knn.score(X_test, y_test)
```

```python
#define my own k-fold cross validation function
from sklearn.cross_validation import KFold
def crossValidate(X, y, classifier, k_folds) :
    
    # get train and test indexes
    indicies = KFold(len(X), n_folds=k_folds, shuffle = True)
    
    #get scores
    scores = []
    for train, test in indicies:
        knn = classifier.fit(X[train],y[train])
        score = knn.score(X[test], y[test])
        scores.append(score)

    # return the average accuracy
    return sum(scores)/float(k_folds)
```

```python
#run my k-fold function on iris data for 5 folds and 3 neighbors
knn = KNeighborsClassifier(3)
crossValidate(X,y,knn,5)
```

```python
#run my k-fold function on iris data for 5 folds and 1-150 neighbors
crossVals = [crossValidate(X,y,KNeighborsClassifier(i),5) for i in range(1,151)]
crossVals
```

```python
#find optimal number of neighbors
crossVals.index(max(crossVals)) + 1
```

```python
#plot hyperparameter k
%matplotlib inline
from matplotlib import pyplot as plt

plt.xlabel('k-neighbors')
plt.ylabel('model accuracy')
plt.scatter(range(1,151), crossVals)
plt.show()
```

```python
#the peak acuracy is between 1 and 20 neighbors so let's look closer at that section
plt.xlabel('k-neighbors')
plt.ylabel('model accuracy')
plt.scatter(range(1,21), crossVals[0:20])
plt.show()
```

We can see that the optimal number of neighbors is 8, after which the accuracy of the k nearest neighbors algorithm starts to decline. The accuracy of the algorithm takes a sharp dip at around 75 neighbors.

```python
#for 11 neighbors, examine a spread of folds
folds = [2, 3, 5, 6, 10, 15]
knn11 = KNeighborsClassifier(8)
vals = [crossValidate(X,y,knn11,i) for i in folds]
for fold, val in zip(folds, vals):
    print 'folds: '+ str(fold), ', accuracy: '+ str(val)
```

The accuracy seems to be roughly the same for each tested number of folds.
