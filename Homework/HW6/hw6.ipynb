## 1) Explore Data

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('bank.csv', sep = ';')
df = data
```

## 2) Encode Labels

```python
from sklearn import preprocessing

job_le = preprocessing.LabelEncoder().fit(df['job'])
df['job'] = job_le.transform(df['job'])

marital_le = preprocessing.LabelEncoder().fit(df['marital'])
df['marital'] = marital_le.transform(df['marital'])

education_le = preprocessing.LabelEncoder().fit(df['education'])
df['education'] = education_le.transform(df['education'])

default_le = preprocessing.LabelEncoder().fit(df['default'])
df['default'] = default_le.transform(df['default'])

housing_le = preprocessing.LabelEncoder().fit(df['housing'])
df['housing'] = housing_le.transform(df['housing'])

loan_le = preprocessing.LabelEncoder().fit(df['loan'])
df['loan'] = loan_le.transform(df['loan'])

contact_le = preprocessing.LabelEncoder().fit(df['contact'])
df['contact'] = contact_le.transform(df['contact'])

poutcome_le = preprocessing.LabelEncoder().fit(df['poutcome'])
df['poutcome'] = poutcome_le.transform(df['poutcome'])

y_le = preprocessing.LabelEncoder().fit(df['y'])
df['y'] = y_le.transform(df['y'])
```

```python
df.head()
```

## 3) Build Decision Tree

```python
from sklearn import tree
model = tree.DecisionTreeClassifier(random_state=0)
features = ['age', 'job', 'marital', 'education', 'default', 'housing',
              'loan', 'contact', 'duration', 'campaign', 'pdays', 'previous']
model.fit(df[features], df['y'])
```

```python
model.score(df[features], df['y'])
```

## 4) Impliment Cross Validation

```python
from sklearn.cross_validation import KFold
def crossValidate(X, y, classifier, k_folds) :

    # get train and test indexes
    indicies = KFold(len(X), n_folds=k_folds, shuffle = True)
    
    #get scores
    scores = 0
    for train, test in indicies:
        model = classifier.fit(X[train],y[train])
        score = model.score(X[test], y[test])
        scores+=score

    # return the average accuracy
    return scores/float(k_folds)
```

```python
crossValidate(df[features].values, df['y'].values, model, 10)
```
